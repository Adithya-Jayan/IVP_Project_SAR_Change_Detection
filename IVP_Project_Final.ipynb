## Importing Required Libraries

```python
import os
import matplotlib.pyplot as plt
import cv2
import glob
import numpy as np
import quaternion
import tqdm
import bispy as bsp
from sklearn.cluster import KMeans
from sklearn.linear_model import LogisticRegression
import time
```

# Step-by-Step Explanation 

#### Read Input Data

```python
img1 = Images[2]
img2 = Images[3]
GroundTruth = Images[4]
if(np.max(GroundTruth)>100):
    GroundTruth = GroundTruth/255


plt.imshow(img1,cmap = 'gray')
plt.title("First Image")
plt.figure()
plt.imshow(img2,cmap = 'gray')
plt.title("Second Image")
```

#### Calculate Difference Image

```python
# Calculate Log difference Image
a = np.double(img1) + 1
b = np.double(img2) + 1

log_diff =np.abs(np.log(np.divide(b,a)))

plt.imshow(log_diff,cmap = 'gray')
plt.title('Log Difference Image')
```

#### Perform Frequency Analysis

```python
r = log_diff/3
g = log_diff/3
b = log_diff/3

I = (r+g+b)/3
R = r-(g+b)/2;
G = g-(r+b)/2;
B = b-(r+g)/2;
Y = (r+g)/2 - np.abs(r-g)/2 -b;
Y[Y<0] = 0
RG = np.double(R - G)
BY = np.double(B - Y)
```

```python
temp = np.stack((RG,BY,I,np.zeros(I.shape)),axis = -1)
Q = quaternion.as_quat_array(temp)
Q.shape
```

```python
Q_fft = bsp.qfft.Qfft(Q)
A = np.abs(Q_fft)
Phase = Q_fft/A

A=np.log(1+np.fft.fftshift(A))

#Kernal size(Tunable Parameter, Good value is 0.05 times Image dimention)
k = 15
k2 = 15

Ak = cv2.GaussianBlur(A, (k,k),cv2.BORDER_REFLECT)
Ak= np.exp(np.fft.ifftshift(Ak))-1
FL_Filtered = Ak*Phase
FIL = bsp.qfft.iQfft(FL_Filtered)
FIL = np.square(np.abs(FIL))

FIL_smoothened = cv2.GaussianBlur(FIL, (k2,k2),cv2.BORDER_REFLECT)
```

```python
size = (8,8)
plt.figure()#figsize=size)
plt.imshow(log_diff,'gray')
plt.title("Log Difference Image")

plt.figure()#figsize=size)
plt.imshow(FIL_smoothened,'gray')
plt.title("After Saliency detection")
```

#### Thresholding generated saliency map using threshold t (Manually set here) and obtain change candidates

```python
# Generate Change Candidates using Thresholding
t = 0.06 #(Tunable Parameter)
Change_candidates = FIL_smoothened.copy()
Change_candidates[Change_candidates>=t] = 1
Change_candidates[Change_candidates<t] = 0

plt.figure()#figsize=size)
plt.imshow(Change_candidates,'gray')
```

```python
Candidates = Change_candidates*log_diff
plt.imshow(Candidates,'gray')
```

#### Segmentation Using K-means Clustering

```python
kmeans = KMeans(n_clusters=3, random_state=0).fit(Candidates.flatten().reshape(-1, 1))

plt.figure(figsize = (10,10))
Labeled = kmeans.labels_
plt.imshow(Labeled.reshape(Candidates.shape),'gray')
print("3 separate Segments - Changed, Unchanged and Unsure\n",np.unique(Labeled))
```

#### Patch Extraction

```python
def PatchExtract(img1,img2):
    #Pad Image
    Padded1 = np.zeros((img1.shape[0]+2,img1.shape[1]+2))
    Padded1[1:-1,1:-1] = img1
    
    Padded2 = np.zeros((img2.shape[0]+2,img2.shape[1]+2))
    Padded2[1:-1,1:-1] = img2
    
    #Extract Patches
    Patches1 = []
    Patches2 = []
    
    time.sleep(0.2)
    for i in tqdm.tqdm(range(img1.shape[0]), desc = "Patch Extraction", position=0, leave=True):
        for j in range(img1.shape[1]):
            Patches1 = Patches1 + [Padded1[i:i+3,j:j+3].flatten()]
            Patches2 = Patches2 + [Padded2[i:i+3,j:j+3].flatten()]
    time.sleep(0.2)
        
    Patches1 = np.array(Patches1)
    Patches2 = np.array(Patches2)
    
    Combined_Patches  = np.concatenate((Patches1,Patches2),axis = 1)
    
    return Combined_Patches
```

```python
Combined_Patches = PatchExtract(img1,img2) ##Takes about 2 minutes
```

#### Uncertain pixel estimation using Logistic regression

```python
clf = LogisticRegression(random_state=0,max_iter=2500,penalty = 'l2').fit(Combined_Patches[np.logical_or(Labeled==0,Labeled==1)], Labeled[np.logical_or(Labeled==0,Labeled==1)])
prediction = clf.predict(Combined_Patches)
output = Labeled.copy()
output[output == 2] = prediction[output == 2]
```

```python
result = output.reshape(img1.shape)

plt.figure(figsize = (15,15))

plt.subplot(1, 3, 1)
plt.imshow(Change_candidates,'gray')
plt.title("Change Candidates")

plt.subplot(1, 3, 2)
plt.imshow(result,'gray')
plt.title("Generated Difference Image")

plt.subplot(1, 3, 3)
plt.imshow(GroundTruth,'gray')
plt.title("Ground Truth")  

plt.show()
```

#### Define Functions to calculate Score metrics

```python
def OveralError(im1,im2):
    diff = np.longlong(np.abs(im1-im2))
    result = np.sum(diff,dtype = np.longlong)
    return result

def PCC(im1,im2):
    OE = OveralError(im1,im2)
    tot = np.longlong(im1.shape[0])*np.longlong(im1.shape[1])
    result = ((tot - OE)/tot)*100
    return result

def Kappa(im1,im2):
    tot = np.longlong(im1.shape[0])*np.longlong(im1.shape[1])
    TP = np.sum(np.logical_and(im1 == 1, im2 == 1))/np.longfloat(tot)
    FP = np.sum(np.logical_and(im1 == 0, im2 == 1))/np.longfloat(tot)
    TN = np.sum(np.logical_and(im1 == 0, im2 == 0))/np.longfloat(tot)
    FN = np.sum(np.logical_and(im1 == 1, im2 == 0))/np.longfloat(tot)
    
    
    pre = ((TP+FP)*(TP+FN) + (FN+TN)*(FP+TN))
    pcc = PCC(im1,im2)/100
    result = (pcc-pre)/(1-pre) *100
    return (result)
```

#### Print metrics

```python
print("OveralError: ", OveralError(result,GroundTruth))
print("Percentage of Correct Classification: ",PCC(result,GroundTruth))
print("Kappa coefficient: ",Kappa(result,GroundTruth))
```

## Implement the entire process as Functions - For ease of applying on Datasets

```python
def ChangeDetection(im1, im2, Patches, t = 0.05):
    
    #Difference image
    
    a = np.double(img1) + 1
    b = np.double(img2) + 1
    log_diff =np.abs(np.log(np.divide(b,a)))
    
    #Freq domain analysis
    
    r = log_diff/3
    g = log_diff/3
    b = log_diff/3
    I = (r+g+b)/3
    R = r-(g+b)/2;
    G = g-(r+b)/2;
    B = b-(r+g)/2;
    Y = (r+g)/2 - np.abs(r-g)/2 -b;
    Y[Y<0] = 0
    RG = np.double(R - G)
    BY = np.double(B - Y)
    
    temp = np.stack((RG,BY,I,np.zeros(I.shape)),axis = -1)
    Q = quaternion.as_quat_array(temp)
    
    #FFT
    Q_fft = bsp.qfft.Qfft(Q)
    A = np.abs(Q_fft)
    Phase = Q_fft/A
    A=np.log(1+np.fft.fftshift(A))
    
    #Calculate k1 and k2
    k1 = np.int32((im1.shape[0]*0.05)//1)
    k2 = np.int32((im1.shape[1]*0.05)//1)
    if(k1%2==0):
        k1 = k1+1
    if(k2%2==0):
        k2 = k2+1
    
    #Amplitude Convolution
    Ak = cv2.GaussianBlur(A, (k1,k2),cv2.BORDER_REFLECT)
    Ak= np.exp(np.fft.ifftshift(Ak))-1
    FL_Filtered = Ak*Phase
    
    #IFFT
    FIL = bsp.qfft.iQfft(FL_Filtered)
    FIL = np.square(np.abs(FIL))
    
    #Smoothening
    FIL_smoothened = cv2.GaussianBlur(FIL, (k1,k2),cv2.BORDER_REFLECT)
    
    #Thersholding
    Change_candidates = FIL_smoothened.copy()
    Change_candidates[Change_candidates>=t] = 1
    Change_candidates[Change_candidates<t] = 0
    
    #Obtain Change Candidates
    Candidates = Change_candidates*log_diff
    
    #3 Level K-means Segmentation
    kmeans = KMeans(n_clusters=3, random_state=0).fit(Candidates.flatten().reshape(-1, 1))
    Labeled = kmeans.labels_

    #Uncertain area prediction
    clf = LogisticRegression(random_state=0,max_iter=2500,penalty = 'l2').fit(Patches[np.logical_or(Labeled==0,Labeled==1)], Labeled[np.logical_or(Labeled==0,Labeled==1)])
    prediction = clf.predict(Patches)
    output = Labeled.copy()
    output[output == 2] = prediction[output == 2]
    
    #Obtain Final Result
    result = output.reshape(img1.shape)
    
    return result
```

```python
def test(im1,im2,gt):
    
    print("Extracting Patches")
    Patches = PatchExtract(img1,img2)
    
    result_values = []
    print("Calculating Results")
    
    time.sleep(0.2)
    
    for t in tqdm.tqdm(np.linspace(0.02,0.2,40),desc="Result Calculation", position=0, leave=True):
        Result = ChangeDetection(im1, im2, Patches, t = t)
        pcc = PCC(Result,gt)
        result_values = result_values + [pcc]
       
    time.sleep(0.2)
    
    print("Plotting results")
    plt.figure()
    plt.plot(np.linspace(0.02,0.2,40),result_values,'g-o')
    plt.title("PCC vs Threshold plot")
    plt.xlabel("t (threshold)")
    plt.ylabel("PCC (%)")
    plt.show()
    
    #Calculate best result
    t = np.linspace(0.02,0.2,40)[result_values == max(result_values)]
    Result = ChangeDetection(im1, im2, Patches, t = t)
    
    print("Best PCC for t = {} :".format(t))
    print("OveralError(OE): ", OveralError(Result,gt))
    print("Percentage of Correct Classification(PCC): ",PCC(Result,gt))
    print("Kappa coefficient: ",Kappa(Result,gt))
    
    plt.figure(figsize = (12,12))
    
    plt.subplot(2, 2, 1)
    plt.imshow(im1,'gray')
    plt.title("Orginal Image 1")
    
    plt.subplot(2, 2, 2)
    plt.imshow(im2,'gray')
    plt.title("Orginal Image 2") 
    
    plt.subplot(2, 2, 3)
    plt.imshow(Result,'gray')
    plt.title("Generated Difference Image")
    
    plt.subplot(2, 2, 4)
    plt.imshow(gt,'gray')
    plt.title("Ground Truth")  
    
    plt.show()
    
```
